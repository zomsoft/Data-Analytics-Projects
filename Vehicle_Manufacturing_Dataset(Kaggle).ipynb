**5.3 - Statistical Metrics Computation**

```python
# Computing metrics like mean, median, Price by grouping by Brand, Model, Year
summary_stats = cars.groupby(['Brand', 'Model', 'Year']).agg({
    'Price': ['mean', 'median', 'min', 'max']
}).reset_index()
summary_stats
```

**Creating a def statement with input function that will Compute the Mean, Median, Minimum and Maximum(2 decimal places) and accepting arguments - Brand, Model, Year**

```python
def car_price_stats(brand, model, year):
    """
    Computes and prints the mean, median, minimum, and maximum prices for a specific car.

    Args:
        brand (str): The brand of the car (e.g., 'Toyota').
        model (str): The model of the car (e.g., 'Corolla').
        year (int): The year of manufacture of the car (e.g., 2020).
    """

    # Convert inputs to lowercase for case-insensitivity
    brand = brand.lower()
    model = model.lower()

    # Filter data (case-insensitive)
    filtered_data = cars[
        (cars['Brand'].str.lower() == brand)
        & (cars['Model'].str.lower() == model)
        & (cars['Year'] == year)
    ]

    if filtered_data.empty:
        print("No matching data found for the specified car. Please try again.")
        return  # Return to input prompt if data is empty

    # Calculate statistics
    mean_price = filtered_data['Price'].mean()
    median_price = filtered_data['Price'].median()
    min_price = filtered_data['Price'].min()
    max_price = filtered_data['Price'].max()

    # Print results formatted to 2 decimal places
    print(f"Price Statistics for {brand.title()} {model.title()} ({year}):")
    print(f"  Mean: ${mean_price:.2f}")
    print(f"  Median: ${median_price:.2f}")
    print(f"  Minimum: ${min_price:.2f}")
    print(f"  Maximum: ${max_price:.2f}")


# Get input from the user, with validation and case-insensitivity
while True:
    brand = input("Enter the brand of the car: ")
    model = input("Enter the model of the car: ")
    try:
        year = int(input("Enter the year of manufacture: "))
    except ValueError:
        print("Invalid year. Please enter a valid integer.")
        continue

    # Call the function to calculate and print statistics
    car_price_stats(brand, model, year)
    break  # Exit the loop if the input is valid
```

**Now let's introduce the Mileage and categorize the values into ranges**

```python
# Calculate mileage range for each Brand, Model, and Year
mileage_range = cars.groupby(['Brand', 'Model', 'Year'])['Mileage'].agg(['min', 'max']).reset_index()
mileage_range['Mileage_Range'] = mileage_range.apply(lambda row: str(int(row['min'])) + 'km to ' + str(int(row['max'])) + 'km', axis=1)  #Creating the range string
mileage_range = mileage_range[['Brand', 'Model', 'Year', 'Mileage_Range']] # Keep relevant columns


# Merge mileage range back into the main DataFrame
cars_with_mileage_range = pd.merge(cars, mileage_range, on=['Brand', 'Model', 'Year'])

# Compute metrics using mileage range
summary_stats = cars_with_mileage_range.groupby(['Brand', 'Model', 'Year', 'Mileage_Range'])['Price'].agg(['mean', 'median', 'min', 'max']).reset_index()
summary_stats = summary_stats.rename(columns={'mean': 'Average Price'})
summary_stats
```

**Creating a def statement that will generate a bar chart for the given brand and model, showing Price vs. Year with Mileage displayed on hover.**



```python
def plot_price_vs_year_with_mileage(brand, model):
    """
    Generates a bar chart for the given brand and model, showing Price vs. Year with Mileage displayed on hover.
    Is case-insensitive and re-prompts for input if invalid data is entered.
    """

    # Convert inputs to lowercase for case-insensitivity
    brand = brand.lower()
    model = model.lower()

    # Filter data (case-insensitive)
    filtered_data = cars[
        (cars['Brand'].str.lower() == brand) & (cars['Model'].str.lower() == model)
    ]

    if filtered_data.empty:
        print("No matching data found for the specified car. Please try again.")
        # Return to input prompt if data is empty
        return False  # Indicate invalid input
    else:
        # ... (Rest of the code to create the bar chart remains the same) ...
        fig = px.bar(
            filtered_data,
            x='Year',
            y='Price',
            title=f'{brand} {model} Price vs Year (Mileage on hover)',  # Dynamic title
            labels={'Price': 'Price', 'Year': 'Year'},
            hover_data=['Mileage'],
            color='Year',
        )

        fig.update_traces(hovertemplate='Year: %{x}<br>Price: %{y}<br>Mileage: %{customdata[0]}')
        fig.update_layout(xaxis_tickangle=0)
        fig.show()
        return True  # Indicate valid input


# Get input from the user, with validation and case-insensitivity
while True:
    brand = input("Enter the brand of the car: ")
    model = input("Enter the model of the car: ")

    # Call the function to generate the chart and validate input
    if plot_price_vs_year_with_mileage(brand, model):
        break  # Exit the loop if the input is valid
```
